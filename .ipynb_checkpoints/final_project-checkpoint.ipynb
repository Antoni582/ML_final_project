## Introduction
### What is the machine learning problem you are trying to solve? 

One primary issue with today's dating apps is the criteria upon which they show a user possible matches. We aim to resolve this problem by analyzing more attributes outside of physical traits and using this data to improve a user's selection of potential matches. The goal of this is to increase the probability that two matched users will have an overall better experience and thus go on a second date.

Therefore, we would want to analyze more attributes of a user's dating preferences to make an optimal matching against a potential candidate.

### What could the result of your predictive model be used for?

The results could be used to further personalize matching suggestions for users based on commonalities between each candidate. This will ideally create a better dating experience.

# Data Cleaning

As the dataset contains 8000+ records, we must explore the data and find any records that have missing or incomplete values.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
data = pd.read_csv("speeddating.csv", header='infer')
print(data.shape)
data.drop_duplicates(inplace=True)
print(data.shape)
data
```

Before performing any data cleaning, we decided to select features from the dataset we considered the most meaningful. This makes cleaning our data more efficient as we do not need to spend time cleaning features we will not be analyzing.

We dropped the "has_null", "wave", "expected_happy_with_sd_people", "expected_num_interested_in_me", "expected_num_matches", "guess_prob_liked", "met", and "field"  features as we considered it irrelevant to our analysis.

We dropped the "sports", "tvsports", "exercise", "dining", "museums", "art", "hiking", "gaming", "clubbing", "reading", "tv", "theater", "movies", "concerts", "music", "shopping", and "yoga" features as all these features are encompassed by the "interests_correlate" feature.

We dropped the "decision" and "decision_o" features because they are encompassed by the class label.

```python
cols_to_drop = [col for col in data.columns if col.startswith("d_")]
data = data.drop(columns = cols_to_drop)
print(data.shape)
cols_to_drop = ["has_null","wave","sports","tvsports","exercise","dining","museums","art","hiking","gaming","clubbing","reading","tv","theater","movies","concerts","music","shopping","yoga", "expected_happy_with_sd_people","expected_num_interested_in_me","expected_num_matches", "guess_prob_liked", "met", "field", "decision", "decision_o"]
data = data.drop(columns = cols_to_drop)
print(data.shape)
print(data.columns)
data
```

We then find which columns contain missing values.

```python
# count null values in each col
null_columns = data.isnull().sum()

cols_with_null_values = null_columns[null_columns > 0].index.tolist()

print(f"Total headers with missing values: {len(cols_with_null_values)} ")

print()
print("Names of headers with missing values:")
print()
for i, col in enumerate(null_columns):
    print(f"Total missing values: {col}. Header name: {data.columns[i]}")
```

There are 35 columns with missing values. Based on our interpretations of which features we considered the most important, we selected those we consider the most significant. For records with missing values, we drop them if their feature values are too subjective to be imputed, otherwise we impute them using appropriate strategies.

We also clean every column whose values contain a "b'" at the beginning of each string. To make our data more readable, we strip these extra characters from every record in the column.

```python
remove_prefix = lambda x: x.strip("b'")

data.loc[:,'gender'] = data['gender'].astype(str).apply(remove_prefix)
for col in data.columns:
    if type(data.iloc[1][col]) == str:
        data.loc[:,col] = data[col].apply(remove_prefix)


data_types_dict = {'samerace': int, 'match':int}
data = data.astype(data_types_dict)
```

We thought that all attributes other than age and age_o are too specific to each record to be imputed. These attributes are highly subjective to the participant, so it is not reasonable to impute using a mean or median. Therefore, we will drop every record with missing values aside from the age and age_o columns. We decided to impute values after our train-test split to avoid data leakage.

```python
col_we_want_to_drop = [header for header in list(data.columns) if header not in ['age', 'age_o']]

data = data.dropna(subset = col_we_want_to_drop)
data 
```

 After dropping records we still have 5803 records left, reducing the dataset by about 30%.

We will then one-hot encode the categorical features race, race_o, and gender.

```python
data = pd.get_dummies(data = data, columns = ["race", "race_o"])
data = pd.get_dummies(data = data, columns = ['gender'], drop_first = True)
data
```

# Data Exploration

We then display the frequency count of matches and find that there is a high imbalance trending towards participants not matching.

```python
print(data['match'].value_counts())
plt.bar(data['match'].unique(), data['match'].value_counts())
```

# Training Data

After sufficient data cleaning and feature engineering, we will start training different models on our data. We chose to keep outliers as some features are highly subjective.

We will be performing 5 different algorithms - 4 supervised, 1 unsupervised. They include: Naive Bayes, K-Means, KNN, Random Forest, and Neural Net

## Naive Bayes Model

The first model we decided to look at is the Naive Bayes model, which is effective at avoiding the curse of dimensionality. As a result, there is no need to scale the data or use PCA. 

This model is highly applicable to our data due to our high number of attributes. 

From data exploration, we noticed that there is a significant class imbalance issue. In attempt to resolve this, we will implement oversampling through SMOTE.

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn import metrics
import numpy as np
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score
from imblearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.impute import SimpleImputer


naive_bayes_data = data

class_labels = naive_bayes_data['match']
attributes = naive_bayes_data.drop(['match'],axis=1)
naive_bayes = GaussianNB()

smote = SMOTE()
pipe = Pipeline(steps=[('imputer', SimpleImputer(strategy = 'median')),('smote', smote), ('GaussianNB',  GaussianNB())])


scores = cross_val_score(pipe, attributes, class_labels, cv=10)       
print("Accuracy:", scores.mean()*100)

X_train, X_test, y_train, y_test = train_test_split(attributes, class_labels, test_size=0.20)
imputer = SimpleImputer(strategy='median')

imputer = imputer.fit(X_train)
X_train = imputer.transform(X_train)
X_test = imputer.transform(X_test)
X_new, y_new = smote.fit_resample(X_train, y_train)
clf = GaussianNB()
clf = clf.fit(X_new, y_new)
pred = clf.predict(X_test)

conf_matrix = confusion_matrix(y_test, pred)
print(conf_matrix)
print(classification_report(y_test, pred))


fpr, tpr = [], []
clf = GaussianNB()
clf = clf.fit(X_new, y_new)
pred = clf.predict_proba(X_test)
fpr, tpr, thresholds = metrics.roc_curve(y_test, pred[:, 1])

plt.plot([0,1],[0,1],'k--') 
plt.plot(fpr, tpr, label='NB') 
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('ROC Curve Naive Bayes')
plt.show()

print(roc_auc_score(y_test, pred[:, 1]))
```

Although there is a decrease in accuracy after implementing SMOTE, we believe this score is a more accurate reflection of our model's performance. The lower accuracy could have been caused by the class imbalance found during data exploration, as this imbalance is still present in the test set.

To improve our Naive Bayes model, we want to get rid of features that are strongly correlated (|r| > 0.6) to preserve independence between attributes.

```python
from numpy import argsort
pd.options.display.max_columns = None
correlation = naive_bayes_data.corr().abs()

unstacked_corr = correlation.unstack()
sorted_corr = unstacked_corr.sort_values(kind="quicksort", ascending=False)
sorted_corr = sorted_corr[sorted_corr.to_frame()[0] < 1]
sorted_corr = sorted_corr[sorted_corr.to_frame()[0] >= 0.6]
sorted_corr
```

Looking at our correlation scores, we noticed that there is a high correlation between the like attribute and the funny_partner, shared_interests_partner, and attractive_partner attributes. 

We decided to remove these attributes because we feel that the "like" attribute is more important than the other three. We did not remove any other attributes because we felt they were equally significant to our dataset, and that their high correlation was coincidental. 

```python
naive_bayes_data.drop(columns=['funny_partner', 'shared_interests_partner', 'attractive_partner'], inplace=True)
naive_bayes_data
```

After dropping funny_partner, shared_interests_partner, and attractive_partner from our Naive Bayes dataset, we run our Naive Bayes model again to check if our model's performance has improved.

```python
class_labels = naive_bayes_data['match']
attributes = naive_bayes_data.drop(['match'],axis=1)
naive_bayes = GaussianNB()

smote = SMOTE()
pipe = Pipeline(steps=[('imputer', SimpleImputer(strategy = 'median')),('smote', smote), ('GaussianNB',  GaussianNB())])


scores = cross_val_score(pipe, attributes, class_labels, cv=10)       
print("Accuracy:", scores.mean()*100)

X_train, X_test, y_train, y_test = train_test_split(attributes, class_labels, test_size=0.20)
imputer = SimpleImputer(strategy='median')

imputer = imputer.fit(X_train)
X_train = imputer.transform(X_train)
X_test = imputer.transform(X_test)
X_new, y_new = smote.fit_resample(X_train, y_train)
clf = GaussianNB()
clf = clf.fit(X_new, y_new)
pred = clf.predict(X_test)

conf_matrix = confusion_matrix(y_test, pred)
print(conf_matrix)
print(classification_report(y_test, pred))


fpr, tpr = [], []
clf = GaussianNB()
clf = clf.fit(X_new, y_new)
pred = clf.predict_proba(X_test)
fpr, tpr, thresholds = metrics.roc_curve(y_test, pred[:, 1])

plt.plot([0,1],[0,1],'k--') 
plt.plot(fpr, tpr, label='NB') 
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('ROC Curve Naive Bayes')
plt.show()

print(roc_auc_score(y_test, pred[:, 1]))
```

Across multiple runs after removing values with high correlation, we see inconsistent changes to the area beneath our ROC curve. Some runs yield an increase in area (evidence of better performance) while others yield a decrease.

## K-means

We use a K-means model to see if our data has any clustering tendencies. This type of model is simple and efficient and can find any clusters that may be present in our data. If we discover any clusters, we could use this information to evalaute closely clustered datapoints, as this may allow us to predict potential matches that exhibit similar qualities.

There is no need to manually scale the data as sci-kit learn will take care of any necessary data scaling. 

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
from warnings import simplefilter

#ignore FutureWarning
simplefilter(action='ignore', category=FutureWarning)

k_means_data = data
class_labels = k_means_data['match']
attributes = k_means_data.drop(['match'],axis=1)

smote = SMOTE()
X_train, X_test, y_train, y_test = train_test_split(attributes, class_labels, test_size=0.20)
imputer = SimpleImputer(strategy='median')

imputer = imputer.fit(X_train)
X_train = imputer.transform(X_train)
X_test = imputer.transform(X_test)
X_new, y_new = smote.fit_resample(X_train, y_train)
scaler = StandardScaler()
X_new = scaler.fit_transform(X_new)
pca = PCA()
X_new = pca.fit_transform(X_new)
X_new = pd.DataFrame(X_new)

sse = []
for k in range(1, 15):
    k_means = KMeans(n_clusters = k).fit(X_new)
    sse.append(k_means.inertia_)
plt.figure()
plt.plot(list(range(1,15)), sse)
plt.show()
    
# Based on our interpretation of the graph, we select 2, 11, 13 as the candidates for the best k value. 
# Next, we will compare their silhouette scores

best_score = -2
best_predicted_labels = None
best_k = 0


for k in [2, 11, 13]:
    predicted_cluster_label = KMeans(n_clusters=k, n_init = 10).fit_predict(X_new)
    score = silhouette_score(X_new, predicted_cluster_label)
    if best_score < score:
        best_score = score
        best_k = k
        best_predicted_labels = predicted_cluster_label

print(best_k)
print(best_score)

# We are trying to use a new centroid initialization method to see if it improves the performance of our model
sse = []
for k in range(1, 15):
    k_means = KMeans(n_clusters = k, init = "k-means++", n_init = 10).fit(X_new)
    sse.append(k_means.inertia_)
plt.figure()
plt.plot(list(range(1,15)), sse)
plt.show()

# Based on our interpretation of the graph, we select 2, 10 as the candidates for the best k value. 
# Next, we will compare their silhouette scores

for k in [2, 10]:
    predicted_cluster_label = KMeans(n_clusters=k, n_init = 10).fit_predict(X_new)
    score = silhouette_score(X_new, predicted_cluster_label)
    if best_score < score:
        best_score = score
        best_k = k
        best_predicted_labels = predicted_cluster_label

print(best_k)
print(best_score)


```

The elbow diagram's large SSE value can be attributed to the vast number of records in our dataset as well as the large amount of features present.

We tried running the K-means algorithm with hyperparameter tuning using different centroid initialization. After tuning hyperparameter K by comparing their silhouette scores, we found that the best K for both initialization methods is 2. However, the silhouette scores were close to 0 (indicating that the points could be clustered in either cluster), meaning the model generated by the K-means clustering algorithm has a poor performance. We think the performance of the model was negatively impacted by the specificity / subjectivity of the attributes. In comparison, Naive Bayes performed much better. 

# KNN

KNN can be an incredibly efficient model, as it is simple and can easily return a high accuracy. However, classification time for test data can be incredibly costly. Feature selection and selecting the best K is essential in creating an accurate model for our data. For this model, we will attempt to find the best number of neighbors to use for classifying the test data.

```python
from sklearn.neighbors import KNeighborsClassifier

# Finding best K
param_grid = {'n_neighbors': np.arange(1, 51)}
knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, param_grid, cv=5)
# We use the same X_new and y_new values we used in K-means as there is no change in the data
grid_search.fit(X_new, y_new)

# Best parameter and score
print("Best parameter: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
    
```

Based on the results from the grid search, the best K is equal to 2. We can see that KNN is effective in predicting the outcomes of speed dating.

Now, we will use our fitted model on the test data to evaluate its performance.

```python
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors=2).fit(X_new, y_new)
predictions = knn.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f"Accuracy using k = 2: {accuracy}")
```

Like Naive Bayes, KNN performed very well in predicting the likelihood of the matches for different pairings. So far, we have seen that supervised learning algorithms are more suited for this data set.

# Random Forest Classifier

By using ensembles of multiple independent base classifiers, we can improve the accuracy of predictions. Here, data doesn't need to be scaled because it's just using decision trees with labels. For this model, we will try to find the best parameters for max depth, number of estimators, min_samples_leaf, and impurity criterions using grid search.

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
forest_data = data
class_labels = forest_data['match']
attributes = forest_data.drop(['match'],axis=1)

X_train, X_test, y_train, y_test = train_test_split(attributes, class_labels, test_size=0.20)
pipe = Pipeline(steps=[(('imputer', SimpleImputer(strategy = 'median'))),('smote', smote), ('RandomForestClassifier', rfc)])

param_grid = {
    'RandomForestClassifier__n_estimators': [50, 100, 150],
     'RandomForestClassifier__max_depth': list(range(2, 10, 5)),
     'RandomForestClassifier__criterion': ['gini', 'entropy'],
     'RandomForestClassifier__min_samples_leaf':list(range(2, 6))
}
grid_search = GridSearchCV(pipe, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
print("Best parameter: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
```

```python
predictions = cross_val_predict(grid_search, attributes, class_labels, cv=5)
print(accuracy_score(class_labels, predictions))
print(classification_report(class_labels, predictions))
```

Random Forest performed well when predicting potential matches.

# Neural Network

Neural networks, despite being time consuming in the training process, can be very effective in discovering various decision boundaries for large amounts of data with many features. Not only can the resulting model be more generalizable, it is also able to assign / rank the importance of attributes based on the weight assignments. For this model, we are trying to find the best hidden layer size, activation function, and learning rate technique using grid search

```python
from sklearn.neural_network import MLPClassifier
from sklearn.exceptions import ConvergenceWarning
simplefilter(action='ignore', category=ConvergenceWarning)

neural_data = data
class_labels = neural_data['match']
attributes = neural_data.drop(['match'],axis=1)
pipe = Pipeline(steps=[(('imputer', SimpleImputer(strategy = 'median'))), ('smote', smote), ('scaler', scaler), ('pca', pca), ('mlp', MLPClassifier())])

param_grid = {
    'mlp__hidden_layer_sizes': list(range(10, 81, 10)),
    'mlp__activation': ['identity','logistic', 'tanh', 'relu'],
    'mlp__learning_rate': ['constant', 'invscaling', 'adaptive']
}
grid_search = GridSearchCV(pipe, param_grid, cv=5, scoring='accuracy')
nested_score = cross_val_score(grid_search, attributes, class_labels, cv=5)
grid_search.fit(X_train, y_train)
print("Best parameter: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
print("Accuracy:", nested_score.mean()*100)
```

We will now run our fitted model on the test set.

```python
imputer = SimpleImputer(strategy='median')

imputer = imputer.fit(X_train)
X_train = imputer.transform(X_train)
X_test = imputer.transform(X_test)
neural_net = MLPClassifier(activation="logistic", hidden_layer_sizes=60, learning_rate="constant").fit(X_train, y_train)
predictions = neural_net.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f"Accuracy using logistic activation function, hidden layer size 60, and constant learning rate: {accuracy*100}")
```

We found that our neural net model has a high accuracy and performs well on the dataset.
